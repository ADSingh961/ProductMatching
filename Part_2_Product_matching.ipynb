Loading the Amazon Dataset

```python
data_amz = pd.read_csv('/content/sample_data/amz_com-ecommerce_sample.csv', encoding='latin1')
data_amz
```

```python
data_amz.columns
```

```python
# Deleting all the columns except the required ones
data_amz.drop(data_amz.columns.difference(['product_name', 'retail_price', 'discounted_price','pid']), 1, inplace=True)
```

```python
data_amz.head()
```

Merging both the dataframes through PID

---



```python
df_matched= pd.merge(df_fk, data_amz, on="pid")
df_matched
```

```python
#Deleting product id for the final result
df_matched.drop('pid',axis=1,inplace=True)
```

```python
# Changing name of all the columns for better understanding
df_matched= df_matched.rename(columns={'product_name_x': 'Product name in Flipkart',
                    'retail_price_x': 'Retail price in Flipkart',
                     'discounted_price_x': 'Discounted price in Flipkart',
                    'product_name_y': 'Product name in Amazon',
                    'retail_price_y': 'Retail price in Amazon',
                     'discounted_price_y': 'Discounted price in Amazon',})

```
